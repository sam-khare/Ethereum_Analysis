```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PartA1.txt', sep='\t', names=['Month-Year', 'Trnx_count'])
df = df.sort_values(by='Month-Year')
df['Month-Year'] = pd.Categorical(df['Month-Year'], ordered=True)
```

```python
df
```

```python
plt.figure(figsize = (22,10))
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Total Transactions')
plt.bar(df['Month-Year'],  df['Trnx_count'])
```
