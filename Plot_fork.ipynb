```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time 
import datetime 

df = pd.read_csv('fork_output.txt',header=None)

new = df[0].str.split("\t", n = 1, expand = True) 
df["day"]= new[0]
df["count"] = new[1]
df["value"] = df[1]
df.drop(columns =[0], inplace = True) 
df.drop(columns =[1], inplace = True) 
df['count'] = df['count'].str.strip('[')
df['value'] = df['value'].str.strip(']')
df
```

```python
df['count'] = df['count'].astype(float)
df['value'] = df['value'].astype(float)
df['day'] = df['day'].astype(int)
df.sort_values(by=['day'],inplace=True)
```

```python
plt.figure(figsize=(20,10))
plt.title('Gas_price vs Day')
plt.xlabel('Day')
plt.ylabel('Gas_price')
plt.yticks(rotation=90)
plt.xticks(rotation=90)
plt.plot(df['day'], df['value'])
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.title('Transaction vs Day')
plt.xlabel('Day')
plt.ylabel('Trnx_count')
plt.yticks(rotation=90)
plt.xticks(rotation=90)
plt.plot(df['day'], df['count'])
plt.show()
```
