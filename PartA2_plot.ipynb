```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PartA2.txt', sep='\t', names=['Month-Year', 'Avg_value'])
df = df.sort_values(by='Month-Year')
df['Month-Year'] = pd.Categorical(df['Month-Year'], ordered=True)
```

```python
df
```

```python
plt.figure(figsize = (25,15))
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Average_value')
plt.bar(df['Month-Year'],  df['Avg_value'])
```
